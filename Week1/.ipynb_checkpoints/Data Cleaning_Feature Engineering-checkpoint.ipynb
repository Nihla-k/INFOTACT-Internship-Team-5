# Data Cleaning, Feature Engineering & Outlier Handling

## 1.Load Dataset

```python
import pandas as pd
import numpy as np


import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('Week1/H1.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

## 2. Handle 'NULL' Values & Missing Value Treatment

```python
# Strip spaces
df = df.apply(lambda x: x.str.strip() if x.dtype == "object" else x)

# Replace NULL-like values
null_values = ['NULL', 'null', 'NaN', 'nan', 'None', '']
df = df.replace(null_values, np.nan)
```

```python
df.isnull().sum()
```

```python
df['Country']=df['Country'].fillna('unknown')
```

```python
df.isnull().sum()
```

```python
df.drop(['Agent','Company'],axis=1,inplace=True)
```

```python
df.isnull().sum()
```

## 3. Remove Duplicates

```python
df.duplicated().sum()
```

```python
df=df.drop_duplicates()
```

```python
df.duplicated().sum()
```

## 4.Standardize Column Names

```python
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
df.head()
```

```python
df.info()
```

## 5. Feature Engineering

```python
df['total_guests'] = df['adults'] + df['children'] + df['babies']
```

```python
df['total_stay'] = df['staysinweekendnights'] + df['staysinweeknights']
```

```python
df['revenue'] = df['adr'] * df['total_stay']
```

```python
month_map = {
"January":1,"February":2,"March":3,"April":4,"May":5,"June":6,
"July":7,"August":8,"September":9,"October":10,"November":11,"December":12
}

df["arrival_month_number"] = df["arrivaldatemonth"].map(month_map)
```

```python
df.isnull().sum()
```

```python
total_revenue=df['revenue'].sum()
print("Total_Revenue:",total_revenue)
```
